```python
import tarfile
import hdf5_getters as getters
import tables
import tempfile
import os
import numpy as np
import threading
import re
```

```python
def generate_analysis(file, file_to_write):    
    file_to_write.write(getters.get_song_id(file).decode('utf-8') + ';')
    file_to_write.write(getters.get_song_hotttnesss(file).astype(str) + ';')
    file_to_write.write(getters.get_analysis_sample_rate(file).astype(str) + ';')
    file_to_write.write(getters.get_danceability(file).astype(str) + ';')
    file_to_write.write(getters.get_duration(file).astype(str) + ';')
    file_to_write.write(getters.get_end_of_fade_in(file).astype(str) + ';')
    file_to_write.write(getters.get_energy(file).astype(str) + ';')
    file_to_write.write(getters.get_key(file).astype(str) + ';')
    file_to_write.write(getters.get_key_confidence(file).astype(str) + ';')
    file_to_write.write(getters.get_loudness(file).astype(str) + ';')
    file_to_write.write(getters.get_mode(file).astype(str) + ';')
    file_to_write.write(getters.get_mode_confidence(file).astype(str) + ';')
    file_to_write.write(getters.get_start_of_fade_out(file).astype(str) + ';')
    file_to_write.write(getters.get_tempo(file).astype(str) + ';')
    file_to_write.write(getters.get_time_signature(file).astype(str) + ';')
    file_to_write.write(getters.get_time_signature_confidence(file).astype(str) + ';')
    file_to_write.write(getters.get_track_id(file).decode('utf-8') + ';')

    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_segments_start(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr + ';')

    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_segments_confidence(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr + ';')

    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_segments_pitches(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr + ';')

    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_segments_timbre(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr + ';')

    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_segments_loudness_max(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr + ';')

    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_segments_loudness_max_time(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr + ';')

    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_segments_loudness_start(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr + ';')

    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_sections_start(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr + ';')

    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_sections_confidence(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr + ';')

    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_beats_start(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr + ';')

    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_beats_confidence(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr + ';')

    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_bars_start(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr + ';')

    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_bars_confidence(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr + ';')

    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_tatums_start(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr + ';')

    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_tatums_confidence(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr)
    file_to_write.write('\n')

def generate_metadata(file, file_to_write):
    file_to_write.write(getters.get_artist_familiarity(file).astype(str) + ';')
    file_to_write.write(getters.get_artist_hotttnesss(file).astype(str) + ';')
    file_to_write.write(getters.get_artist_id(file).decode('utf-8') + ';')
    file_to_write.write(getters.get_artist_latitude(file).astype(str) + ';')
    file_to_write.write(getters.get_artist_longitude(file).astype(str) + ';')
    file_to_write.write("'" + getters.get_artist_location(file).decode('utf-8') + '\';')
    file_to_write.write(getters.get_artist_name(file).decode('utf-8') + ';')
    file_to_write.write(getters.get_song_id(file).decode('utf-8') + ';')
    file_to_write.write(getters.get_title(file).decode('utf-8') + ';')
    
    arr = ' '.join(["'" + x.decode('utf-8') + "'" for x in getters.get_similar_artists(file)])
    file_to_write.write(arr + ';')
    
    arr = ' '.join(["'" + x.decode('utf-8') + "'" for x in getters.get_artist_terms(file)])
    file_to_write.write(arr + ';')
    
    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_artist_terms_freq(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr + ';')
    
    arr = re.sub(r'[\n\t\s]+', ' ', np.array2string(getters.get_artist_terms_weight(file).flatten(), precision=3, separator=' ', threshold=np.inf))
    file_to_write.write(arr + ';')
    
    file_to_write.write(getters.get_year(file).astype(str))
    file_to_write.write('\n')
```

```python
analysis_file = open('data/analysis.txt', 'w')
metadata_file = open('data/metadata.txt', 'w')

with tarfile.open('test.tar.gz', 'r:gz') as tar:
    for member in tar.getmembers():
        if member.isfile():
            file_content = tar.extractfile(member).read()
            
            temp_file = tempfile.NamedTemporaryFile(suffix='.h5', delete=False)
            temp_file.write(file_content)
            temp_file_name = temp_file.name
            
            h5file = tables.open_file(temp_file_name)
            try:
                t1 = threading.Thread(target=generate_analysis, args=(h5file, analysis_file))
                t2 = threading.Thread(target=generate_metadata, args=(h5file, metadata_file))
                
                t1.start(); t2.start()
                t1.join(); t2.join()
            finally:
                h5file.close()
            
            temp_file.close()
            os.remove(temp_file_name)

analysis_file.close()
metadata_file.close()
print('Finished')
```
